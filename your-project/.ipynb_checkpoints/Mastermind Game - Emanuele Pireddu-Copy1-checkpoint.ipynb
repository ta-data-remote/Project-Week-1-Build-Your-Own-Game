## How to play


The purpose of the game is to _deduce_ the secret code, achieved through a selection of colored balls.

The difficulty changes the length of the secret sequence.


***
**Easy** = 4 balls<br>
**Normal** = 6 balls<br>
**Hard** = 8 balls<br>

***











After entering all needed values, the game gives out 2 hints:
* Number of balls answered that match the correct position.
* Number of balls answered that are actually in the secret sequence but are not in order.

```python



########## GAME METHODS

def input_difficulty_check(): #Difficulty selector and checker
    
    while True: 
        dif = input('1 = Easy mode\n2 = Normal mode\n3 = Hard mode\nPlease select difficulty! (Enter 1, 2 or 3)  ' )

        try:
            dif = int(dif) #Assign the type int to input i, although float should be an option too (?)
            if dif == 1 or dif == 2 or dif == 3:
                break

            else:
                continue
        except:
            print('Please enter a valid number (1, 2 or 3)') 
            continue
        break
    
    return dif
 
def difficulty_selection(x):
    if x == 1:
        balls_length = 4 #instead of putting -1 in the for len() it is already the needed length
        print('Easy mode selected')
    elif x == 2:
        balls_length = 6
        print('Normal mode selected')
    elif x == 3:
        balls_length = 8
        print('Hard mode selected')
   
    return balls_length



def randomizer_sequence(number_balls, options_balls): #THE MOST DIFFICULT SO FAR
    list1 = []
    for i in range(0, number_balls):
        list1.extend(options_balls[r.randint(0, len(options_balls) - 1 )]) #OMG THE PROBLEMS I HAD till... added - 1 to the len(opt balls) cuz everytime random was 8 out of index
 
    return list1


def needed_inputs(correct_sequence, balls): #nº of inputs = nº values in correct_sequence
    input_list = [] #where we store the user's choices ATTENTION it must be len as len correct sequence in other words, code starts when all inputs are entered.
    answer = ''
    
    while len(correct_sequence) != len(input_list) or answer not in balls:
        
        answer = input('\nPlease choose a ball color (R, B, G, Y, O, W, P, C):')
   
        if answer in balls: #CHECK IF ANSWER IS IN BALLS (LIST) HOW TO COMPARE X WITH VALUES IN LIST
            input_list.append(answer)
                       
        else:
            
            print('Please enter a valid color:') 
        
        print('Your choices are', input_list)    
        continue
    return input_list 
 
   
def check_answer(user_input, correct_sequence):
    counter_correct_position = 0
    counter_correct_color = 0
    correct_answer = True
   
    for i in range(0, len(user_input)):                        ##### I believe that maybe with list comprehension this could be achieved in a faster/shorter way. (?)
        if user_input[i] == correct_sequence[i]:
            counter_correct_position += 1
        else:
            correct_answer = False
            
            if user_input[i] in correct_sequence and user_input[i] != correct_sequence[i]: # ATTENTION CANT ACTUALLY COUNT IF INPUT IS IN C_SEQUENCE BUT NOT REPEATING
            
                    counter_correct_color += 1

    

    # [R,Y,G,B]
       # [G,Y,B,Y]
       # Posibilidades:
       # 0:Color no en la lista  
       # 1:Color en la lista pero no en su sitio
       # 2:Color en su sitio
        #[1,2,1,1]
        

    if correct_answer:
        print('You win!')
        play_again = True
        while play_again: 
            replay = input('Want to play again? Y/N:')

            try:
                if replay == 'Y':
                    play_game()

                elif replay == 'N':
                    play_again = False 
            except:
                break
    else:
        print('Wrong, please try again.')
        print(counter_correct_position, "balls are in the correct position.") 
        print(counter_correct_color, "balls have been correctly chosen.")
        
    return correct_answer
                           
def play_game():
    
    ##### GAME INFO
    difficulty = input_difficulty_check()
    balls_length = difficulty_selection(difficulty)
    correct_sequence = randomizer_sequence(balls_length, balls)
    print(correct_sequence)
    user_input = needed_inputs(correct_sequence, balls)

    while not check_answer(user_input, correct_sequence):
        user_input = needed_inputs(correct_sequence, balls)


```
